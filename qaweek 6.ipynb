# 3.countries with respect to the region

```python
countries .groupby("Region")[["country"]].count()
```

# 4.a. region with highest GDP per capita

```python
Top_GDPMAX=countries.groupby("Region")[["GDP per capita (current US$)"]].mean()
Top_GDPMAX.sort_values(by="GDP per capita (current US$)",ascending=False)
```

```python
Top_GDPMAX.nlargest(1,"GDP per capita (current US$)")
```

# 4.b. Region with highest import

```python
Top_ImpMAX=countries.groupby("Region")[["International trade: Imports (million US$)"]].mean()
Top_ImpMAX.sort_values(by="International trade: Imports (million US$)",ascending=False)
```

```python
Top_ImpMAX.nlargest(1,"International trade: Imports (million US$)")
```

# 5.Region has more consistent with respect to GDP per capita

```python
GP=countries.groupby("Region")[["GDP per capita (current US$)"]].std()
GP.sort_values(by="GDP per capita (current US$)")
```

```python
GP.nsmallest(1,"GDP per capita (current US$)")
```
